```python
import os
from PIL import Image
import tifffile

# 定義要處理的資料夾
tiff_folder = r"C:\Users\anywhere4090\Desktop\ggmfg\Dynamic-noise-AD-master\dataset\mvtec3d\potato(xyz)"

# 遞迴遍歷資料夾，尋找所有 TIFF 文件
for root, dirs, files in os.walk(tiff_folder):
    for file in files:
        if file.lower().endswith('.tiff'):
            tiff_image_path = os.path.join(root, file)
            png_image_path = os.path.join(root, file.replace('.tiff', '.png'))
            
            # 使用 tifffile 讀取 TIFF 文件
            img = tifffile.imread(tiff_image_path)
            
            # 將浮點數數據轉換為整數類型（如果需要）
            img = (img * 255).astype('uint8')
            
            # 使用 PIL 儲存為 PNG
            pil_img = Image.fromarray(img)
            pil_img.save(png_image_path, "PNG")
            print(f"已將 {tiff_image_path} 轉換為 PNG，保存為：{png_image_path}")
```
